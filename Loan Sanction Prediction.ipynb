```python
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split,cross_validate 
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_features,target,test_size=0.2
                                                    ,random_state=40)
```

```python
cat_cols = train_features.select_dtypes(include='object').columns
```

```python
cat_cols
```

```python
models = [LGBMRegressor(random_state=40),XGBRegressor(random_state=40),ExtraTreesRegressor(random_state=40),LinearRegression(),CatBoostRegressor(random_state=40,verbose=0)]
for model in models:
    print(f"********************{model}*********************")
    cv = cross_validate(model,train_features,target,cv=10,scoring=['r2'])
    print(cv['test_r2'].mean())
```

```python
selected_columns = ['Loan Amount Request (USD)','Credit Score','Co-Applicant','Income Stability','No. of Defaults','Income (USD)','Profession','Has Active Credit Card']
for col in train_features.drop(selected_columns,axis=1).columns:
    print(f"***************{col}*******************")
    cv = cross_validate(LGBMRegressor(random_state=40),np.log1p(train_features[[col]+selected_columns]),target,cv=10,scoring=['r2'])
    print(cv['test_r2'].mean())
```
